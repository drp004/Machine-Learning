```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Live.csv")
df.head()
```

```python
df.drop(['status_id', 'status_published', 'Column1', 'Column2', 'Column3', 'Column4'], axis='columns', inplace=True)
df.head()
```

```python
df.describe()
```

```python
plt.bar(df['status_type'], df['num_reactions'])
plt.xlabel("Status Type")
plt.ylabel("Number of Reactions")
plt.show()
```

```python
plt.bar(df['status_type'], df['num_loves'])
plt.xlabel("Status Type")
plt.ylabel("Number of Loves")
plt.show()
```

```python
plt.bar(df['status_type'], df['num_wows'])
plt.xlabel("Status Type")
plt.ylabel("Number of Loves")
plt.show()
```

```python
df1 = df[df['status_type'] == "video"]
df2 = df[df['status_type'] == "photo"]
df3 = df[df['status_type'] == "link"]
df4 = df[df['status_type'] == "status"]
```

```python
plt.scatter(df1['num_reactions'], df1['num_shares'], color="red")
plt.scatter(df2['num_reactions'], df2['num_shares'], color="green")
plt.scatter(df3['num_reactions'], df3['num_shares'], color="blue")
plt.scatter(df4['num_reactions'], df4['num_shares'], color="yellow")
plt.xlabel("Number of Reactions")
plt.ylabel("Number of Shares")
plt.legend(['video', 'photo', 'link', 'status'])
plt.show()
```
