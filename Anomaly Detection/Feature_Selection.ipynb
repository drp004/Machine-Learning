```python
import matplotlib.pyplot as plt
from scipy.stats import skewnorm

numValues = 10000
maxValues = 100
skewness = 20

random = skewnorm.rvs(a=skewness, loc=maxValues, size=numValues) #Skewnorm function

random = random - min(random)
random = random / max(random)
random = random * maxValues

x = random

plt.hist(x, bins=50, color='r')
```

```python
import numpy as np

plt.hist(np.log(x), bins=50)     # bcz some values are zero so they become infinity.
```

```python
plt.hist(np.log(x+5), bins=50)
```
